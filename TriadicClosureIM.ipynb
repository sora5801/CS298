```python
def complement(u):
  return set(G.nodes) - u
```

```python
def average_degree():
  sum = 0
  for v in G.nodes():
    sum += G.degree[v]
  return sum / G.number_of_nodes()
```

```python
def triadic_closure(u, v): # Number of triadic closures between u and v
  n = 0
  for i in G.nodes:
    if G.has_edge(u, i) and G.has_edge(v, i) and G.has_edge(u, v):
      n += 1
  return n
```

```python
def td(u): # Triadic Closure Weighted Degree
  sum = 0
  for v in (set(G.neighbors(u))).union({u}):
    sum += (1 + omega * triadic_closure(u, v))
  return sum
```

```python
def p(u, v):  # Triadic Closure Weighted Propagation Probability
  sum = 0
  for k in G.neighbors(v):
    sum += td(k)
  return td(u) / sum
```

```python
def utd(u, S):  # Updated triadic closure weighted degree
  sum = 0
  temp_set = set()
  for g in S:
    for x in G.neighbors(g):
      temp_set = temp_set.union({x})
  temp_set.union(S)
  for v in (set(G.neighbors(u))).union({u}).intersection(complement(temp_set)):
    sum += (1 + omega * triadic_closure(u, v))
  return sum
```

```python
def e(u): # Triadic Closure Weighted Expected Influence
  sum = 0
  for v in G.neighbors(u):
    sum += p(u, v) * td(v)
  return sum + 1
```

```python
def eUpdate(u, S):# Triadic Closure Weighted Expected Influence updating
  sum1 = 0
  for v in set(G.neighbors(u)).difference(S):
    sum1 += p(u, v) * utd(v, S)
  sum2 = 0
  for j in set(G.neighbors(u)).intersection(S):
    sum3 = 0
    for v in set(G.neighbors(j)).intersection(set(G.neighbors(u)).difference(S)):
      sum3 += p(u, v) * p(j, v) * utd(v, S)
    sum2 += sum3
  return 1 + sum1 - sum2
```

```python
def argmax(d, S):
  max = float('-inf')
  item = 0
  for x in d:
    if d[x] > max and x not in S:
      max = d[x]
      item = x
  return item
```

```python
def tcim(G, k): # Triadic Closure-based Influence Maximization
  updated = {}
  global S1
  for v in G.nodes:
    updated[v] = 0
  td_table = {}
  e_table = {}
  for v in G.nodes:
    td_table[v] = td(v)
  for v in G.nodes:
    e_table[v] = e(v)
  while len(S1) < k:
    u = argmax(e_table, S1)
    if updated[u] == 1:
      S1 = S1.union({u})
      e_table[u] = float('-inf')
      for v in (set(G.nodes) - S1):
        updated[v] = 0
    else:
      for j in (set(G.neighbors(u)) - S1):
        td_table[j] = utd(j, S1)
      e_table[u] = eUpdate(u, S1)
      updated[u] = 1
  return S1
```

```python
def set_node_community(G, communities):
    '''Add community to node attributes'''
    for c, v_c in enumerate(communities):
        for v in v_c:
            # Add 1 to save 0 for external edges
            G.nodes[v]['community'] = c + 1

def set_edge_community(G):
    '''Find internal edges and add their community to their attributes'''
    for v, w, in G.edges:
        if G.nodes[v]['community'] == G.nodes[w]['community']:
            # Internal edge, mark with community
            G.edges[v, w]['community'] = G.nodes[v]['community']
        else:
            # External edge, mark as 0
            G.edges[v, w]['community'] = 0
```

```python
def get_color(i, r_off=1, g_off=1, b_off=1):
    r0, g0, b0 = 0, 0, 0
    n = 16
    low, high = 0.1, 0.9
    span = high - low
    r = low + span * (((i + r_off) * 3) % n) / (n - 1)
    g = low + span * (((i + g_off) * 5) % n) / (n - 1)
    b = low + span * (((i + b_off) * 7) % n) / (n - 1)
    return (r, g, b)
```

```python
# Set node and edge communities
set_node_community(G, partition)
set_edge_community(G)

# Set community color for nodes
node_color = [get_color(G.nodes[v]['community']) for v in G.nodes]

# Set community color for internal edges
external = [(v, w) for v, w in G.edges if G.edges[v, w]['community'] == 0]
internal = [(v, w) for v, w in G.edges if G.edges[v, w]['community'] > 0]
internal_color = [get_color(G.edges[e]['community']) for e in internal]
```

```python
# Modified tcim to take into account communities
def tcim_c(G, c, k):
  for i in c:
    global S2
    updated = {}
    S_i = set()
    for v in G.nodes:
      if v in i:
        updated[v] = 0
    td_table = {}
    e_table = {}
    for v in G.nodes:
      if v in i:
        td_table[v] = td(v)
    for v in G.nodes:
      if v in i:
        e_table[v] = e(v)
    while len(S_i) < k:
      u = argmax(e_table, S2)
      if updated[u] == 1:
        S_i = S_i.union({u})
        e_table[u] = float('-inf')
        for v in (set(G.nodes) - S_i):
          if v in i:
            updated[v] = 0
      else:
        for j in (set(G.neighbors(u)) - S_i):
          if v in i:
            td_table[j] = utd(j, S2)
        e_table[u] = eUpdate(u, S2)
        updated[u] = 1
    S2 = S2.union(S_i)
  return S2
```

```python
def can_reach(G, u, v): # Use basic BFS to determine if there is a path between node u and v.
  queue = []
  queue.append(u)
  visited = []
  visited.append(u)
  while queue:
    s = queue.pop(0)
    for nodes in G.neighbors(s):
      if nodes not in visited:
        queue.append(nodes)
        visited.append(s)
  if v in visited:
    return True
  else:
    return False
```

We will now change the formulas to count the number of Jaccard Coefficients of each node to see how the propagation probability and expected influence changes.

```python
def complement(u):
  return set(G.nodes) - u
```

```python
def average_degree():
  sum = 0
  for v in G.nodes():
    sum += G.degree[v]
  return sum / G.number_of_nodes()
```

```python
S1 = set()
omega = average_degree()
```

```python
def jd(u): # Jaccard Weighted Degree
  sum = 0
  for v in (set(G.neighbors(u))).union({u}):
    sum += (1 + omega * list(nx.jaccard_coefficient(G, [(u, v)]))[0][2])
  return sum
```

```python
def jp(u, v):  # Jaccard Weighted Propagation Probability
  sum = 0
  for k in G.neighbors(v):
    sum += jd(k)
  return jd(u) / sum
```

```python
def uj(u, S):  # Jaccard weighted degree
  sum = 0
  temp_set = set()
  for g in S:
    for x in G.neighbors(g):
      temp_set = temp_set.union({x})
  temp_set.union(S)
  for v in (set(G.neighbors(u))).union({u}).intersection(complement(temp_set)):
    sum += (1 + omega * list(nx.jaccard_coefficient(G, [(u, v)]))[0][2])
  return sum
```

```python
def ej(u): # Jaccard Weighted Expected Influence
  sum = 0
  for v in G.neighbors(u):
    sum += jp(u, v) * jd(v)
  return sum + 1
```

```python
def ejUpdate(u, S):# Jaccard Weighted Expected Influence updating
  sum1 = 0
  for v in set(G.neighbors(u)).difference(S):
    sum1 += jp(u, v) * uj(v, S)
  sum2 = 0
  for j in set(G.neighbors(u)).intersection(S):
    sum3 = 0
    for v in set(G.neighbors(j)).intersection(set(G.neighbors(u)).difference(S)):
      sum3 += jp(u, v) * jp(j, v) * uj(v, S)
    sum2 += sum3
  return 1 + sum1 - sum2
```

```python
def argmax(d, S):
  max = float('-inf')
  item = 0
  for x in d:
    if d[x] > max and x not in S:
      max = d[x]
      item = x
  return item
```

```python
def jim(G, k): # Jaccard-based Influence Maximization
  updated = {}
  global S1
  for v in G.nodes:
    updated[v] = 0
  td_table = {}
  e_table = {}
  for v in G.nodes:
    td_table[v] = jd(v)
  for v in G.nodes:
    e_table[v] = ej(v)
  while len(S1) < k:
    u = argmax(e_table, S1)
    if updated[u] == 1:
      S1 = S1.union({u})
      e_table[u] = float('-inf')
      for v in (set(G.nodes) - S1):
        updated[v] = 0
    else:
      for j in (set(G.neighbors(u)) - S1):
        td_table[j] = uj(j, S1)
      e_table[u] = ejUpdate(u, S1)
      updated[u] = 1
  return S1
```

```python
def triadic_closure_c(u, v, c): # Number of triadic closures between u and v while i is a triadic closure in the same community as u and v.
  n = 0
  for i in G.nodes:
    if G.has_edge(u, i) and G.has_edge(v, i) and G.has_edge(u, v) and i in c and v in c:
      n += 1
  return n
```

```python
def tdc(u, c): # triadic closure community Weighted Degree
  sum = 0
  for v in (set(G.neighbors(u))).union({u}):
    sum += (1 + omega * triadic_closure_c(u,v,c))
  return sum
```

```python
def tdpc(u, v, c):  # Triadic Closure Propagation Probability
  sum = 0
  for k in G.neighbors(v):
    sum += tdc(k, c)
  return tdc(u, c) / sum
```

```python
def utdc(u, S, c):  # Triadic closure weighted degree
  sum = 0
  temp_set = set()
  for g in S:
    for x in G.neighbors(g):
      temp_set = temp_set.union({x})
  temp_set.union(S)
  for v in (set(G.neighbors(u))).union({u}).intersection(complement(temp_set)):
    sum += (1 + omega * triadic_closure_c(u,v,c))
  return sum
```

```python
def etdc(u, c): # triadic closure Expected Influence with community
  sum = 0
  for v in G.neighbors(u):
    sum += tdpc(u, v, c) * tdc(v, c)
  return sum + 1
```

```python
def ejUpdatec(u, S, c):# Triadic Closure Expected Influence updating with community
  sum1 = 0
  for v in set(G.neighbors(u)).difference(S):
    sum1 += tdpc(u, v, c) * utdc(v, S, c)
  sum2 = 0
  for j in set(G.neighbors(u)).intersection(S):
    sum3 = 0
    for v in set(G.neighbors(j)).intersection(set(G.neighbors(u)).difference(S)):
      sum3 += tdpc(u, v, c) * tdpc(j, v, c) * utdc(v, S, c)
    sum2 += sum3
  return 1 + sum1 - sum2
```

```python
def argmax(d, S):
  max = float('-inf')
  item = 0
  for x in d:
    if d[x] > max and x not in S:
      max = d[x]
      item = x
  return item
```

```python
def timc(G, k, c): # Triadic Closure Influence Maximization with community
  updated = {}
  global S1
  for v in G.nodes:
    updated[v] = 0
  td_table = {}
  e_table = {}
  for v in G.nodes:
    td_table[v] = tdc(v, c)
  for v in G.nodes:
    e_table[v] = etdc(v, c)
  while len(S1) < k:
    u = argmax(e_table, S1)
    if updated[u] == 1:
      S1 = S1.union({u})
      e_table[u] = float('-inf')
      for v in (set(G.nodes) - S1):
        updated[v] = 0
    else:
      for j in (set(G.neighbors(u)) - S1):
        td_table[j] = utdc(j, S1, c)
      e_table[u] = ejUpdatec(u, S1, c)
      updated[u] = 1
  return S1
```

Algorithm pseudocode
Given a graph G=(V,E) and k budget
1) Find communities
2) Remove intra edges (edges between communities) -> new graph G’=(V, E’), where E’ contains only inter edges (edges between communities
3) Compute TDI in G’
4) Run QuotaBased influence for each community, i.e. figure out how many of the k to assign to each community.
5) Find the seed nodes of each community put them in set S
6) Use set S in G to find the spread.
